```python
import pandas as pd
```

```python
df = pd.read_csv('weather_smog.csv', parse_dates=True)
df.set_index(['date'], inplace=True)
df.index = pd.to_datetime(df.index)
df.set_index('station', append=True, inplace=True)
df
```

```python
conditions = ['temperature', 'precipitation_1', 'humidity', 'wind']
```

```python
def createDataFrameWithHistoricalDataForParam(df: pd.DataFrame, param, history):
    frame = df[[*conditions, param]].copy(deep=True)
    frame.dropna(inplace=True)
    for hour in history:
        frame[f'{param}-{hour}'] = df.groupby(level=0)[param].shift(hour, freq='H')
    frame.dropna(inplace=True)
    return frame
```

```python
exdf = createDataFrameWithHistoricalDataForParam(df, 'NO2', [2, 4])
```
