```python
import pandas as pd
```

```python
weather = pd.read_excel('wmo_points.xlsx')
smog = pd.read_excel('smog_points.xlsx')
```

```python
weather
```

```python
weather_points = [(el[0], el[1], el[2] + el[3]/60, el[4] + el[5]/60) for i, el in weather.iterrows()]
weather_points
```

```python
smog_points = [(el[1], el[13], el[14]) for i, el in smog.iterrows()]
smog_points
```

```python
def distance(A_y, A_x, B_y, B_x):
    return  (((A_y - B_y) ** 2 + (A_x - B_x) ** 2)**0.5)*111
```

```python
points_pairs = []
for A in weather_points:
    distances = []
    for B in smog_points:
        x = distance(A[2], A[3], B[1], B[2])
        distances += [(B, x)]
    closest = min(distances, key=lambda p: p[1])
    points_pairs.append((A[1],A[0], closest[0][0], closest[1]))
```

```python
points_pairs.sort(key=lambda x: x[3])
for pair in points_pairs:
    print(f'{pair[1]};{pair[0]};{pair[2]};')
```
